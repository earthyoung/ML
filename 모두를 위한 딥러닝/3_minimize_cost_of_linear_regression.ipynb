```python
import tensorflow.compat.v1 as tf1
# 슬프지만 v2 버전으로 어떻게 같은 코드를 구현하는지 몰라서 v1의 tensorflow를 사용하기로 함
import matplotlib.pyplot as plt
tf1.disable_v2_behavior()

X = [1,2,3]
Y = [3,6,9]

# W에다가 다양한 값을 입력해 볼 것이므로 W을 placeholder로 설정한다.
W = tf1.placeholder(tf1.float32)
hypothesis = X * W

cost = tf1.reduce_mean(tf1.square(hypothesis-Y))

#Session 시작하기(with 사용하지 않고 session을 그냥 정의했음)
sess = tf1.Session()

# tf에 있는 global variable들을 initialize하는 작업
sess.run(tf1.global_variables_initializer())    

# cost function을 그리기 위한 variable 선언
wval = []
costval = []
         
for i in range(-30, 50):
    feed_W = i * 0.1
    curr_cost, curr_W = sess.run([cost, W], feed_dict = {W: feed_W})
    wval.append(curr_W)
    costval.append(curr_cost)
                                 
    plt.plot(wval, costval)
    plt.show()
    
    
sess.close()
    
    
    # 밑에 -30부터 50까지 약 80개의 그래프가 나옴.
```

```python

import tensorflow.compat.v1 as tf

# 이번에는 x, y를 소문자로 하겠음
# cost plot에서 W를 update 시키는 과정을 코드로 표현할 것임.

x_data = [1,2,3]
y_data = [1,2,3]

#w, x, y 할당
w = tf1.Variable(tf1.random_normal([1]), name = 'weight')   # random_normal : version1, random.normal : version2!! 이거 헷갈려서 계속 에러나옴ㅠㅠ
x = tf1.placeholder(tf1.float32)
y = tf1.placeholder(tf1.float32)

# hypothesis 설정
hypothesis_new = x * w

# cost 설정
cost_new = tf1.reduce_mean(tf1.square(hypothesis_new-y))

# W의 값을 update하는 식을 만들 때 사용할 변수들 설정
learning_rate = 0.1
gradient = tf1.reduce_mean((x*w-y)*x) # sigma 를 적용하고 평균 내는 것과 같은 원리 : tf.reduce_mean
descent = W - learning_rate * gradient
update = w.assign(descent)   # =을 사용해서 값을 할당할 수 없고, assign이라는 함수/메소드를 사용해야 가능

# Session 시작하기
sess_new = tf1.Session()
sess_new.run(tf1.global_variables_initializer())

# update하고 결과 출력하기
for step in range(21):
    sess_new.run(update, feed_dict = {x: x_data, y: y_data})
    print(step, sess_new.run(cost_new, feed_dict = {x: x_data, y: y_data}), sess_new.run(w))
    
    
sess_new.close()
    
    
```

```python
import tensorflow.compat.v1 as tf

# 이번에는 x, y를 소문자로 하겠음
# cost plot에서 W를 update 시키는 과정을 코드로 표현할 것임.

x2 = [1,2,3]
y2 = [1,2,3]

#w, x, y 할당
w2 = tf1.Variable(5.0)   # random_normal : version1, random.normal : version2!! 이거 헷갈려서 계속 에러나옴ㅠㅠ

# hypothesis 설정
hypothesis2 = x2 * w2

# cost 설정
cost_new2 = tf1.reduce_mean(tf1.square(hypothesis2-y2))

#GradientDescentOptimizer 알고리즘 사용할 것이므로 선언(정의)
optimizer2 = tf1.train.GradientDescentOptimizer(learning_rate=0.1)
train2 = optimizer2.minimize(cost_new2)

# Session 시작하기
sess_new2 = tf1.Session()
sess_new2.run(tf1.global_variables_initializer())

# update하고 결과 출력하기
for step in range(21):
    print(step, sess_new2.run(w2))
    sess_new2.run(train2)
    
    
sess_new2.close()
    
    
```
